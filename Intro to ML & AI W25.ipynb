#### Neural Networks

What you've been waiting for :-)

So, before we get started, what do you guys know about neural networks?

Base Terms: 
- Neural Networks: made up of layers of connected units (neurons) that process / learn info
- Neuron: takes input, processes it, and passes output on
- Input layer: takes in data
- Hidden layers: where data is processed and transformed so the NN can 'learn'
- Output layer: gives prediction

More Mathy: 
- Epoch: one pass through the data (we can train for multiple epochs, meaning pass the data thru NN multiple times)
- Batch Size: subset of the training data used in one iteration of the training process
- Weights: determine the importance of each input; each connection between neurons has a weight that is 'learned' during training
- Activation function: introduces non-linearity, allowing NN to learn complex relationships

Types of NN: 
- Convolutional Neural Networks: specialized NN used primarily for image processing; use layers that apply filters (or convolutions) to input data, helping them detect features like edges, textures, and shapes
- Recurrent Neural Networks: designed for processing sequences of data, like time series or language; RNN's have a 'memory' and can use previous info to influence predictions

- Transfer Learning: use a pretrained model as a base to then train on new data
